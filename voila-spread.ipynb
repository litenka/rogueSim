```python
# Combat roll

def combatrollblanc(totalmiss, DodgeChance, glancing, CritChance, hit ):
    """
    combat roll pour les coups blanc
    
    """
    
    aa_milne_arr = ['miss', 'Dodge', 'Glancing', 'Crit', 'hit']
    a = np.random.choice(aa_milne_arr, 1, p=[totalmiss,DodgeChance ,glancing, CritChance,hit ])
    return a[0]

def combatrollbs(DodgeChance, CritChance, hit ):
    """
    combat roll pour les back stab
    
    """
    aa_milne_arr = ['Dodge', 'Crit', 'hit']
    a = np.random.choice(aa_milne_arr, 1, p=[DodgeChance, CritChance + 0.3, 1- DodgeChance -(CritChance + 0.3 )])

    return a[0]

def combatrollEvi(DodgeChance, CritChance, hit ):
    """
    combat roll pour les evicerates
    
    """
    aa_milne_arr = ['Dodge', 'Crit', 'hit']
    a = np.random.choice(aa_milne_arr, 1, p=[DodgeChance, CritChance, 1- DodgeChance -CritChance ])
    return a[0]

def glancingdmg(weaponskillsss, degats):
    if weaponskillsss >= 308:
        degats = degats*0.95
    elif weaponskillsss == 307:
        degats = degats*0.93
    elif weaponskillsss == 306:
        degats = degats*0.89
    elif weaponskillsss == 305:
        degats = degats*0.85
    elif weaponskillsss == 304:
        degats = degats*0.81

    return degats
```

```python
# Proc talents enchantements stuff

def nemesis(): 
    """
    nemesis; 60% de chances de generer 1 pts de combo apres une attaque finale
    
    """
    aa_milne_arr = ['non', 'oui']
    nem = np.random.choice(aa_milne_arr, 1, p=[0.4,0.6])
    
    return nem[0]

def frappeimp(combo):
    """
    Frappe implacable; 
    1 point de combo, 20% de chance d'avoir 25 energies apres une attaque finale
    2 point de combo, 40% de chance d'avoir 25 energies apres une attaque finale
    3 point de combo, 60% de chance d'avoir 25 energies apres une attaque finale
    4 point de combo, 80% de chance d'avoir 25 energies apres une attaque finale
    5 point de combo, 100% de chance d'avoir 25 energies apres une attaque finale
    
    """
    if combo == 1:
        aa_milne_arr = ['non', 'oui']
        frp = np.random.choice(aa_milne_arr, 1, p=[0.8,0.2])
    elif combo == 2:
        aa_milne_arr = ['non', 'oui']
        frp = np.random.choice(aa_milne_arr, 1, p=[0.6,0.4])
    elif combo == 3:
        aa_milne_arr = ['non', 'oui']
        frp = np.random.choice(aa_milne_arr, 1, p=[0.4,0.6])
    elif combo == 4:
        aa_milne_arr = ['non', 'oui']
        frp = np.random.choice(aa_milne_arr, 1, p=[0.8,0.2])
    elif combo == 5:
        aa_milne_arr = ['non', 'oui']
        frp = np.random.choice(aa_milne_arr, 1, p=[1,0])
    
    return frp[0]



def procpoisonInstant():
    
    aa_milne_arr = ['non', 'oui']
    poison = np.random.choice(aa_milne_arr, 1, p=[0.76,0.24])
    return poison[0]


def poisonInstant():
    
    PIdegatmin = 112
    PIdegatmax = 148
    PIdegats = random.randint(PIdegatmin, PIdegatmax)
    aa_milne_arr = [0, PIdegats, PIdegats*0.25, PIdegats *0.5, PIdegats*0.75]
    poisonok = np.random.choice(aa_milne_arr, 1, p=[0.17,0.55,0.2,0.06,0.02])
    return poisonok[0]
    

def ProcarmureRogue():
    aa_milne_arr = ['non', 'oui']
    bonus = np.random.choice(aa_milne_arr, 1, p=[0.965,0.035])
    
    return bonus[0]
    

def armureRogue(energie):

    energie +=35
    # nombre de proc pour stats

    return energie
    

def proccrusader(speedweapon):
    
    proc = (speedweapon * 1.82)/100
    aa_milne_arr = ["oui", "non"]
    proccroise = np.random.choice(aa_milne_arr, 1, p=[proc, 1 - proc])
    
    return proccroise[0]
    
def crusader(mhdmg_min, mhdmg_max, ohdmg_min, ohdmg_max, speedmh, speedoh, ap, enchantedeg):
    
    crusadertime = 1500
    MHdegatmin = mhdmg_min + enchantedeg + ((speedmh * (ap+100))/14)
    MHdegatmax = mhdmg_max + enchantedeg + ((speedmh* (ap+100))/14)
    OHdegatmin = ohdmg_min +  (speedoh * (ap +100)/14)*0.5
    OHdegatmax = ohdmg_max +  (speedoh * (ap +100) /14)*0.5
    return crusadertime, MHdegatmin, MHdegatmax, OHdegatmin, OHdegatmax

def prochoj():
    
    aa_milne_arr = ["oui", "non"]
    prochoj = np.random.choice(aa_milne_arr, 1, p=[0.02, 0.98])
    return prochoj[0]
```

```python
# Competances

def castAutoAttack(temps, arme, prochaineattaque, degatmin, degatmax, armorreduc, 
           totalmiss, DodgeChance, glancing, CritChance, hit, weaponskill, 
           speed, energie, debite, newspeed):
    
    degats = random.randint(int(degatmin), int(degatmax))
    degats = degats / armorreduc

    f = combatrollblanc(totalmiss, DodgeChance, glancing, CritChance, hit)

    if f =='hit':
        degatactu = degats
    elif f =='miss':
        degatactu = 0
    elif f =='Dodge':
        degatactu = 0
    elif f =='Glancing':
        degatactu = glancingdmg(weaponskill, degats)
    elif f =='Crit':
        degatactu = degats *2

    if temps == 0:
        
        mhspeedactu0 = int(speed*100)
        nextattack = temps + mhspeedactu0
        
    
    elif debite is True:

        nextattack = temps + int(newspeed*100)

    else:
        
        nextattack = temps + int(speed*100)
        

    # proc 35 energie si armure voleur
    procnrj = ProcarmureRogue()

    if procnrj == "oui" and (f != 'Dodge' or f != "miss"):

        energie = armureRogue(energie)

    

    # max energie = 100
    energie = energiemax(energie)


    # Poison instantané
    piproc = procpoisonInstant()
    
    degatpoison = 0
    
    if piproc =='oui' and (f != 'Dodge' and f != 'miss'):
        
        degatpoison = poisonInstant()
    
    return f, degatactu, nextattack, energie, procnrj, piproc, degatpoison, newspeed
    
def castBs(temps, energie, MHdegatmin, MHdegatmax ):
    backstap = random.randint(int(MHdegatmin), int(MHdegatmax)) * 1.5 + 225 * 1.2
    return backstap
    
def castdebite(debite, energie, mhspeed, ohspeed, combo):
    energie -= 25
    debitemhspeed = mhspeed / 1.3
    debiteohspeed = ohspeed / 1.3
    debite = True
    debitetime = 0
    if combo == 1:
        debitetime = 900 *1.45
        combo -=1
    elif combo == 2:
        debitetime = 1200 *1.45
        combo -=2
    elif combo == 3:
        debitetime = 1500 *1.45
        combo -=3
    elif combo == 4:
        debitetime = 1800 *1.45
        combo -=4
    elif combo == 5:
        debitetime = 2100 *1.45
        combo -=5
            
    return debitetime, debitemhspeed, debiteohspeed, debite, combo

def castevi(combo, ap):
    if combo == 4:
        evimin = 652 + (ap *0.15)
        evimax = 748 + (ap *0.15)
        
        evi = random.randint(int(evimin), int(evimax)) *1.15
        
    elif combo == 5:
        
        evimin = 803 + (ap *0.15)
        evimax = 899 + (ap *0.15)
        evi = random.randint(int(evimin), int(evimax)) *1.15
        
    return evi



def combomax(combo):
    if combo > 5:
        combo = 5
    return combo

def energiemax(energie):
    if energie > 100:
        energie = 100
    return energie

```

```python
def miliseconde(millis):
    millis = int(millis)
    millisprint=round((millis/1)%100,2)
    seconds=(millis/100)%60
    seconds = int(seconds)
    minutes=(millis/(100*60))%60
    minutes = int(minutes)
    hours=(millis/(100*60*60))%24

    return ("%d:%d:%d.%d " % (hours, minutes, seconds, millisprint))
```

```python
def journalcombat(g) :
    
    cblanc = g[(g.Action == 'Main Hand') | (g.Action == 'Off Hand')]
    Main = cblanc[cblanc.Action == 'Main Hand']
    Off = cblanc[cblanc.Action == 'Off Hand']
    BackStab = g[g.Action == 'BackStap']
    Evi = g[g.Action == 'Eviscerate']
    
    print ('dps : ', round(g.Degats.sum()/120,2))
    print('proc 35 energie :       ',len(g[g.Action == 'Proc energie']))
    
    
    print('coup blanc :            ', int(len(cblanc)))
    
    print('            - Miss : ', round((len(cblanc[cblanc.Type == 'miss'])/len(cblanc))*100,1))
    print('            - Esquive : ', round((len(cblanc[cblanc.Type == 'Dodge'])/len(cblanc))*100,1))
    print('            - Glancing : ', round((len(cblanc[cblanc.Type == 'Glancing'])/len(cblanc))*100,1))
    print('            - Coup critique : ', round((len(cblanc[cblanc.Type == 'Crit'])/len(cblanc))*100,1))
    print('            - Hit Normal : ', round((len(cblanc[cblanc.Type == 'hit'])/len(cblanc))*100,1))

    
    print ('total Main Hand : ', len(Main))
    print ('total Off Hand : ', len(Off))
    print ('total BackStab : ', len(BackStab))
    print ('total Eviscerate : ', len(Evi))
    print ('----------')

    for i in range(len(g)):

        print ('---------|----------------------------------------------------------')
        if len(g.loc[i+1,'Time']) == 8:
            f = g.loc[i+1,'Time'] +'00  |'
            f = f.replace(' ', '')
            f = f +   str(g.loc[i+1,'Color1']) + str(g.loc[i+1,'Action'])+ str(g.loc[i+1,'Color2'])
            f = f + ' - ' +str(g.loc[i+1,'Type'])
            f = f + ' - ' + str(g.loc[i+1,'Degats'])
            print (f)

        elif len(g.loc[i+1,'Time']) == 9:
            a = g.loc[i+1,'Time'] +'0  |' 
            a = a.replace(' ', '')
            a = a + str(g.loc[i+1,'Color1']) + str(g.loc[i+1,'Action'])+ str(g.loc[i+1,'Color2'])
            a = a + ' - ' +  str(g.loc[i+1,'Type']) 
            a = a + ' - ' + str(g.loc[i+1,'Degats'])
            print (a)

        else :
            z = g.loc[i+1,'Time'] +'  |' 
            z = z.replace(' ', '')
            z = z +  str(g.loc[i+1,'Color1']) + str(g.loc[i+1,'Action']) + str(g.loc[i+1,'Color2'])
            z = z + ' - ' +  str(g.loc[i+1,'Type'])
            z = z + ' - ' + str(g.loc[i+1,'Degats']) 
            print (z)
```
